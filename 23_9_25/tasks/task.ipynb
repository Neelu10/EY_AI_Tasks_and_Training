```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python

# 1. Basic Analysis
# Total revenue per product
revenue_per_product = df.groupby('Product')['Revenue'].sum()
print("Total revenue per product:")
print(revenue_per_product)
```

```python

# Average units sold per product
avg_units_sold_per_product = df.groupby('Product')['Units_Sold'].mean()
print("\nAverage units sold per product:")
print(avg_units_sold_per_product)
```

```python
import matplotlib.pyplot as plt
monthly_revenue =df.groupby(['Product','Month'])['Revenue'].sum().reset_index()
pivot_df=monthly_revenue.pivot(index='Month', columns='Product', values='Revenue')
pivot_df.plot(marker='o')
plt.title('Monthly Revenue trend by product')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(title="Product")
plt.grid()
plt.show()
```

```python

# 3. Visualization 2  Bar Chart: Total revenue per product
plt.figure(figsize=(6, 4))
revenue_per_product.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Total Revenue per Product')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

```python

# 4. Visualization 3  Pie Chart: Revenue share per product
plt.figure(figsize=(6, 6))
revenue_per_product.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Revenue Share per Product')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
